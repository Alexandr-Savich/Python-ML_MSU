```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
urla='https://raw.githubusercontent.com/vaibhavwalvekar/NYC-Flights-2013-Dataset-Analysis/master/flights.csv'
Ar = pd.read_csv(urla)
Ar.head(25)
```

1.Исследуем среднее время задержки отправления и прибытия от перевозчика

```python
Airlines=Ar.groupby('carrier')['arr_delay'].mean()
Airlines.plot.bar()
plt.show()
Airlines=Ar.groupby('carrier')['dep_delay'].mean()
Airlines.plot.bar()
plt.show()
Ar['carrier'].value_counts().plot.bar()
plt.show()
```

По задержкам отправления и прибытия существенно выделяются авиакомпании EV, F9, FL, YV. Все, кроме Ev малого размера, поэтому им нужно лучше оптимизировать распределение ресурсов. Ev же рационально нанять больше персонала и постараться организовать его работу эффективнее, подобно другим большим перевозчикам

4.Зависимость от дней недели

```python
S=np.arange(12)
G=np.zeros(12)
S=Ar.groupby('month')['day'].max()
for i in range(11):
    G[i+1]=G[i]+S[i+1]
#print(G)
Ar['week']=(Ar['day']+G[Ar['month']-1])%7
Ar['days']=Ar['day']+G[Ar['month']-1]
Airlines=Ar.groupby('week')['dep_delay'].mean()
Airlines.plot.pie()
plt.ylabel('Dep_delay',fontsize=13)
plt.show()
Ar['week'].value_counts().plot.bar()
plt.ylabel('Departures',fontsize=13)
plt.show()
```

Как видно из графика, наблюдается определенная неоднородность в эффективности работы аэропорта в разные дни недели (4 и 3 особенно выделяются). Наибольшая средняя задержка наблюдается в 3 день недели, причем загрузка аэропорта по первым 4-ем дням равномерна. Различия в средней задержке могут быть связаны с днями недели - где большая задержка - дни перед выходными, из-за чего работает меньше персонала. Выходом будет стимулирование персонала выходить на работу в эти дни.
